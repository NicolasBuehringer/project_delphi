```python
import twint
import nest_asyncio
import pandas as pd

query = "@AfDimBundestag OR @afdimbundestag OR @AfD OR @afd OR @Alice_Weidel OR @alice_weidel OR @Tino_Chrupalla OR @tino_chrupalla OR #Chrupalla OR #chrupalla OR #Weidel OR #weidel OR #AfD OR #afd"

nest_asyncio.apply()
#configuration
config = twint.Config()
config.Search = query
config.Lang = "de"
config.Since = "2021-07-01"
config.Until = "2021-08-14"
config.Store_csv = True
config.Output = "afd_bis_08_16.csv"
#running search
twint.run.Search(config)
```
