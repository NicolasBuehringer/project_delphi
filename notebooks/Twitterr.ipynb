```python
import pandas as pd
import nest_asyncio
import matplotlib.pyplot as plt
nest_asyncio.apply()
pd.set_option('display.max_colwidth', None)
import twint
```

```python

config = twint.Config()
# config.Lang = "fr"
# config.Min_likes= 10
german_party=["CDU"]
#"SPD", "Gruenen", "FDP", "AfD", "LINKE"
for party in german_party:
    print(f"{party}")
    config.Search = party
    config.Limit = 100_000
    config.Min_likes= 10
    config.Lang = "de"
    config.Since ='2017-08-17 12:00:00'
    config.Until ='2017-08-17 15:00:00'
    config.Store_csv = True
    config.Hide_output = True
    config.Output = f"{party}_german_party.csv"
    twint.run.Search(config)
```
